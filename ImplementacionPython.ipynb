```python
from datetime import datetime

# Definici√≥n de las clases
class Equipo:
    def __init__(self, conductor_nombre, conductor_documento, asistente1_nombre, asistente1_documento, asistente2_nombre, asistente2_documento):
        self.conductor_nombre = conductor_nombre
        self.conductor_documento = conductor_documento
        self.asistente1_nombre = asistente1_nombre
        self.asistente1_documento = asistente1_documento
        self.asistente2_nombre = asistente2_nombre
        self.asistente2_documento = asistente2_documento

class Camion:
    def __init__(self, placa, equipo, ruta, dia):
        self.placa = placa
        self.equipo = equipo
        self.ruta = ruta
        self.dia = dia

class Punto:
    def __init__(self, longitud, latitud):
        self.longitud = longitud
        self.latitud = latitud

class PuntoRecoleccion:
    def __init__(self, puntos, clasificacion):
        self.puntos = puntos
        self.clasificacion = clasificacion

class Ruta:
    def __init__(self, punto_origen, puntos_recoleccion, punto_clasificacion, hora_inicio, hora_fin, dia):
        self.punto_origen = punto_origen
        self.puntos_recoleccion = puntos_recoleccion
        self.punto_clasificacion = punto_clasificacion
        self.hora_inicio = hora_inicio
        self.hora_fin = hora_fin
        self.dia = dia
        self.cantidad_clasificacion = {}

    def actualizar_cantidad_clasificacion(self, turno, clasificacion):
        if turno in self.cantidad_clasificacion:
            self.cantidad_clasificacion[turno].recibir_residuos(clasificacion)
        else:
            self.cantidad_clasificacion[turno] = LugarClasificacion()
            self.cantidad_clasificacion[turno].recibir_residuos(clasificacion)


class Clasificacion:
    def __init__(self, vidrio=0, papel=0, plastico=0, metal=0, organico=0):
        self.vidrio = vidrio
        self.papel = papel
        self.plastico = plastico
        self.metal = metal
        self.organico = organico

class LugarClasificacion:
    _instance = None

    def __new__(cls, *args, **kwargs):
        if not cls._instance:
            cls._instance = super().__new__(cls)
        return cls._instance

    def __init__(self):
        if not hasattr(self, 'clasificacion_total'):
            self.clasificacion_total = Clasificacion()

    def recibir_residuos(self, clasificacion):
        self.clasificacion_total.vidrio += clasificacion.vidrio
        self.clasificacion_total.papel += clasificacion.papel
        self.clasificacion_total.plastico += clasificacion.plastico
        self.clasificacion_total.metal += clasificacion.metal
        self.clasificacion_total.organico += clasificacion.organico


# Plan de pruebas
equipo = Equipo("Conductor", "123456789", "Asistente1", "987654321", "Asistente2", "567891234")
punto_origen = Punto(0.0, 0.0)
puntos_recoleccion = [Punto(1.0, 1.0), Punto(2.0, 2.0), Punto(3.0, 3.0)]
punto_clasificacion = Punto(4.0, 4.0)
hora_inicio = datetime.now()
hora_fin = datetime.now()
dia = "2023-05-19"
clasificacion = Clasificacion(vidrio=10, papel=20, plastico=30, metal=40, organico=50)

ruta = Ruta(punto_origen, puntos_recoleccion, punto_clasificacion, hora_inicio, hora_fin, dia)
ruta.actualizar_cantidad_clasificacion("Turno1", clasificacion)

print(ruta.cantidad_clasificacion)
```

```python
import unittest

class TestLugarClasificacion(unittest.TestCase):
    def test_recibir_residuos(self):
        lugar = LugarClasificacion()
        clasificacion = Clasificacion(vidrio=10, papel=5, plastico=8, metal=3, organico=2)
        
        lugar.recibir_residuos(clasificacion)
        
        self.assertEqual(lugar.clasificacion_total.vidrio, 10)
        self.assertEqual(lugar.clasificacion_total.papel, 5)
        self.assertEqual(lugar.clasificacion_total.plastico, 8)
        self.assertEqual(lugar.clasificacion_total.metal, 3)
        self.assertEqual(lugar.clasificacion_total.organico, 2)
```
